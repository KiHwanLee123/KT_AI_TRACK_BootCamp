##Linear Regression on 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
x = boston.data
y = boston.target

x.shape, y.shape
```

```
y[:3]
```

```
import tensorflow as tf
from tensorflow import keras
import numpy as mp
```

```
keras.backend.clear_session
model = keras.models.Sequential()
model.add(keras.layers.Input(shape=(13,)))
model.add(keras.layers.Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
print(y[:5])
```

```
model.predict(x).reshape(-1)[:5]
```

```
boston = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/boston.csv')
x = boston.iloc[:,:-1]
y = boston.iloc[:,-1]
x.shape, y.shape
```

```
keras.backend.clear_session
model = keras.models.Sequential()
model.add(keras.layers.Input(shape=(12,)))
model.add(keras.layers.Dense(1, activation='linear'))
model.compile(loss=keras.losses.mean_squared_error, optimizer=keras.optimizers.Adam())
```

```
model.fit(x, y, epochs=10, verbose=1)
model.predict(x).reshape(-1)[:5]
```

```
y[:5]
```

## Logistic Regression 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_breast_cancer
```

```
cancer = load_breast_cancer()
```

```
x = cancer.data
y = cancer.target

x.shape, y.shape
```

```
np.unique(y)
```

```
# 이전 모델 없애기
keras.backend.clear_session
# 모델 생성
model = keras.models.Sequential()
# 추가
model.add(keras.layers.Input(shape=(30,)))
model.add(keras.layers.Dense(1, activation='sigmoid'))
# 컴파일
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
model.predict(x).reshape(-1)[:5] # 1 1 0 0 0
```

```
y[:5]
```
