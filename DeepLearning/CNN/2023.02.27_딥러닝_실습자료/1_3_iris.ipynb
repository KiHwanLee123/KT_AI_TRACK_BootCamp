```
from sklearn.datasets import load_iris
```

```
iris = load_iris()
```

```
x = iris.data
y = iris.target
```

```
y
```

```
iris.target_names
```

```
# One-Hot Encoding
from tensorflow.keras.utils import to_categorical
```

```
y = to_categorical(y, 3)
```

```
y[:5]
```

```
x.shape, y.shape
```

## 1번. Sequential API style로 모델링 하세요

- loss = 'categorical_crossentropy' 사용!

```
import tensorflow as tf
from tensorflow import keras

keras.backend.clear_session()
model = keras.models.Sequential()
model.add(keras.layers.Input(shape=(4,)))
model.add(keras.layers.Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
model.predict(x)[:5] # 001 001 001 001 001
```

```
y[:5]
```

## 1번. Sequential API style로 모델링 하세요

- loss = 'sparse_categorical_crossentropy' 사용!

```
import tensorflow as tf
from tensorflow import keras

keras.backend.clear_session()
model = keras.models.Sequential()
model.add(keras.layers.Input(shape=(4,)))
model.add(keras.layers.Dense(3, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

ㅁ ㅁ ㅁ ㅁ

    o  o  o
   로 간다 -> 그 때 추가로 모델을 일반적으로 더 잘 쓸 수 있게 bias를 더해준다

```
model.summary() # parameter : (4+1) * 3 = 15
```

```
model.fit(x, y, epochs=10, verbose=1)
```

```
model.predict(x)[:5] # 001 001 001 001 001
```

```
y[:5]
```
