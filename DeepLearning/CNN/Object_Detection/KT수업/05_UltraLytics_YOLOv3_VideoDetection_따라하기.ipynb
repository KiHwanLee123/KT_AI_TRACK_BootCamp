```python
temp_str = 'setuptoos<=64.0.2\n'

f = open('/content/yolov3/requirements.txt', 'r')
f_str = f.readlines()
f.close()

f2 = open('/content/yolov3/requirements.txt', 'w')

for idx, val in enumerate(f_str):
  if 'setuptoos' in val:
    idx_v = idx
    f_str.remove(val)
    f_str.insert(idx_v, temp_str)
for val in f_str:
  f2.write(val)
f2.close()
```
