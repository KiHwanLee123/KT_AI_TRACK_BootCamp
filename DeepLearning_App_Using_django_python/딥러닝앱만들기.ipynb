```python
import gradio as gr

def greet(name):
  print(name)
  return "Hello " + name + "!"

demo = gr.Interface(fn=greet, inputs="text", outputs="text")

demo.launch(share=True, debug=True)
```

```python
import gradio as gr
ko2enDict = {
    "사과" : "apple",
    "사자" : "tiger",
    "사랑" : "love"
}
def trans_ko_en(ko):
    return ko2enDict[ko]
app = gr.Interface(
    fn=trans_ko_en,
    inputs=gr.Textbox(placeholder="한국어"),
    outputs="text",
)
app.launch(debug=True)
```

```python
import gradio as gr
dict = {
    "ko2en" : {
        "사과" : "apple",
        "사자" : "tiger",
        "사랑" : "love"
    },
    "en2ko" : {
        "apple" : "사과",
        "tiger" : "사자",
        "love" : "사랑"
    }
}
def trans(type, word):
    return dict[type][word]
app = gr.Interface(
    fn=trans,
    inputs=[
        gr.Radio(['ko2en', 'en2ko']),
        gr.Textbox(placeholder="한국어")
    ],
    outputs="text",
)
app.launch(debug=True)

```

```python
import gradio as gr
dict = {
    "ko2en" : {
        "사과" : "apple",
        "사자" : "tiger",
        "사랑" : "love"
    },
    "en2ko" : {
        "apple" : "사과",
        "tiger" : "사자",
        "love" : "사랑"
    }
}
def trans(type, word):
    return dict[type][word]
with gr.Blocks() as app: # UI를 바꿔줄 수 있다.
    result = gr.Textbox(label='결과')
    type = gr.Radio(label='종류', choices=['ko2en', 'en2ko'])
    word = gr.Textbox(label='단어', placeholder="한국어")
    btn = gr.Button(value='실행')
    btn.click(
        fn = trans,
        inputs = [type, word],
        outputs = result
    )
app.launch(debug=True)
```

```python
# blocks

import gradio as gr
from transformers import pipeline

def sentiment(word):
    classifier = pipeline("sentiment-analysis")
    return classifier(word)

with gr.Blocks() as app:
    result = gr.Textbox(label='결과')
    word = gr.Textbox(label='분석할 문장')
    btn = gr.Button(value='실행')

    btn.click(
        fn = sentiment,
        inputs = word,
        outputs = result
    )

app.launch()
```
