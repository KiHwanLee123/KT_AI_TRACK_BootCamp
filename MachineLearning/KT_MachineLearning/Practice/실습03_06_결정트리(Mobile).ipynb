```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/jangrae/csv/master/mobile_cust_churn.csv'
data = pd.read_csv(path)
```

- 분석할 데이터를 충분히 이해할 수 있도록 다양한 탐색 과정을 수행합니다.

# 3.데이터 준비

- 전처리 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

**1) 변수 제거**

```python
# 제거 대상: id
data.drop('id', axis=1, inplace=True)

# 변수 제거


# 확인
data.tail()
```

**2) x, y 분리**

```python
# Target 설정
target='CHURN'

# 데이터 분리
x=data.drop(target, axis=1)
y=data[target]
```

**3) 가변수화**

```python
# 가변수화 대상: 'REPORTED_SATISFACTION, REPORTED_USAGE_LEVEL, CONSIDERING_CHANGE_OF_PLAN]
cols = ['REPORTED_SATISFACTION','REPORTED_USAGE_LEVEL','CONSIDERING_CHANGE_OF_PLAN']

# 가변수화
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in cols:
    x[i] = le.fit_transform(x[i])
# 확인
x.head()
```

```python
x = pd.get_dummies(x,columns=cols)
```

**4) 학습용, 평가용 데이터 분리**

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

# 4.모델링

- 본격적으로 모델을 선언하고 학습하고 평가하는 과정을 진행합니다.

```python
# 1단계: 불러오기
from sklearn.tree import DecisionTreeClassifier
```

```python
# 2단계: 선언하기
model = DecisionTreeClassifier()
```

```python
# 3단계: 학습하기
model.fit(X_train, y_train)
```

```python
# 4단계: 예측하기
y_pred = model.predict(X_test)
```

```python
# 5단계: 평가하기
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# 5.기타

- 기타 필요한 내용이 있으면 진행합니다.

```python
# 트리 시각화
# 시각화 모듈 불러오기
from sklearn.tree import export_graphviz
from IPython.display import Image

# 이미지 파일 만들기
export_graphviz(model,                                 # 모델 이름
                out_file='asdqwe.dot',                   # 파일 이름 
                feature_names=x.columns,               # Feature 이름
                class_names=['LEAVE', 'STAY'],       # Target Class 이름
                rounded=True,                          # 둥근 테두리
                max_depth=3,                           # 출력할 트리 깊이 지정   
                precision=2,                           # 불순도 소숫점 자리수
                filled=True)                           # 박스 내부 채우기

# 파일 변환
!dot asdqwe.dot -Tpng -oasdqwe.png -Gdpi=300

# 이미지 파일 표시
Image(filename='asdqwe.png')
```

```python
# 변수 중요도 시각화
# 변수 중요도
pert_dic = {'feature':list(x), 'importance':model.feature_importances_}
df = pd.DataFrame(pert_dic)
df.sort_values('importance', ascending=True, inplace=True)

plt.figure(figsize=(5, 5))
plt.barh(df['feature'],df['importance'])
plt.show()
```
