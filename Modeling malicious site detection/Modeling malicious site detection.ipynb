```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy

# X, y 데이터 분리
X = X_train.iloc[:,:-1]
y = X_train['label']

# 랜덤 포레스트 모델 생성
rf = RandomForestClassifier(n_estimators=200, n_jobs=-1, class_weight='balanced', max_depth=100)

# Boruta 알고리즘 수행
boruta = BorutaPy(estimator=rf, n_estimators='auto', max_iter=100, verbose=2, random_state=42)
boruta.fit(X.values, y.values)

# 선택된 feature 확인
selected_features = X.columns[boruta.support_]
print(selected_features)
```

```python
len(selected_features)
```

```python
selected_features = ['url_len', 'url_num_hyphens_dom', 'url_path_len', 'url_domain_len',
       'url_hostname_len', 'url_num_dots', 'url_query_len', 'url_entropy',
       "html_num_tags('script')", "html_num_tags('div')",
       "html_num_tags('form')", "html_num_tags('a')"]

selected_features.append('label')
X_train = X_train[selected_features]
```

```python
x = X_train.iloc[:,:-1]
y = X_train['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
y_test.shape
```

```python
from pycaret.classification import *
clf = setup(data=X_train, target=y_train, use_gpu=True)
```

```python
models()
```

```python
best_model = compare_models(n_select=8, include=['lr','knn','nb','svm','rf','xgboost','lightgbm','catboost'])
```

```python
best_model
```

```python
tuned_best_model = tune_model(best_model[0])
```

```python
tuned_best_model1 = tune_model(best_model[1])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
tuned_best_model.fit(X_train, y_train)
```

```python
tuned_best_model1.fit(X_train, y_train)
```

```python
y_pred = tuned_best_model.predict(X_test)

from sklearn.metrics import *
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
y_pred1 = tuned_best_model1.predict(X_test)

from sklearn.metrics import *
print(confusion_matrix(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```
